## Compare the models.

ml33 - LinearRegression.
ml34 - Neural Network.

```python
%matplotlib inline
# Dependencies.
import pandas as pd
from matplotlib import style
style.use('fivethirtyeight')
```

```python
# Read in.
df33 = pd.read_csv('ml_results/33_ml_results.csv')['Difference']
df34 = pd.read_csv('ml_results/34_nn_model.csv')['Difference']
```

```python
# Create comparison.
compare_df = pd.concat([df33, df34], axis=1)
compare_df.columns = ['33_d', '34_d']
compare_df.head()
```

```python
# Plot.
for i in range(0, 200, 40):
    compare_df.iloc[i:i+40].plot(y=['33_d', '34_d'], figsize=(15,3.5))
```
