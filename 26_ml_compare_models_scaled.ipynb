Here we will read in the first two result DataFrames from notebooks 20 and 21, and compare how accurate they were.

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies.
import pandas as pd
```

```python
df_20 = pd.read_csv('ml_results/20_ml_results.csv')
df_20.head()
```

```python
df_25 = pd.read_csv('ml_results/25_ml_results.csv')
df_25.head()
```

```python
# Rename columns.
df_20.columns = ['20_Predicted', '20_Actual', '20_Difference']
df_25.columns = ['25_Predicted', '25_Actual', '25_Difference']
# Merge, and create difference DF.
merged_df = df_20.join(df_25)
diff_df = merged_df[['20_Difference', '25_Difference']]
diff_df.head()
```

```python
# Compare the differences.
diff_df.describe()
```

```python
# Style.
from matplotlib import style
style.use('fivethirtyeight')
```

```python
# Plot the differences.
bins = [0, 40, 80, 120, 160]
for i in bins:
    diff_df.iloc[i:i+40].plot(figsize=(15,3.5))
```

From the above graphs we can see that the data scaling we did, did not affect the model results. It's time to investigate PCA and then a new model.
