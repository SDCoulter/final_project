## Scrape numbeo.com to get a table of QoL ratings.

```python
# Dependencies.
import pandas as pd
```

### Scrape the Table featured on the webpage.

```python
# Read the data from the website, store the SECOND table element as a DataFrame
df = pd.read_html('https://www.numbeo.com/quality-of-life/rankings.jsp')[1]
# Make the index reflect the ranking position, and change the name of the index.
df.index += 1
df.index.name = 'Rank'
# Drop actual 'Rank' column.
df = df.drop(['Rank'], 1)
# Display DataFrame.
df.head()
```

```python
# Save scraping output.
df.to_csv('resources/qol_ratings.csv')
```

Data is scraped from "https://www.numbeo.com/quality-of-life/rankings_current.jsp".
