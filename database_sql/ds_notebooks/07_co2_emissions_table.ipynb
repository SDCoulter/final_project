## Create the Unstacked Version of the CO<sub>2</sub> Emissions Dataset

```python
# Dependencies.
import pandas as pd
```

```python
# Data.
co2_df = pd.read_csv('../co2_emissions_table.csv', index_col=['Country', 'Year'])
co2_df.head()
```

```python
# Unstack and save CSV.
unstack_df = co2_df.unstack()
unstack_df = unstack_df.loc[:, 'CO2 Emissions']
unstack_df.columns.name = None
unstack_df.to_csv('../country_index/co2_emissions_table.csv')
unstack_df.head()
```
