## Create the Clean Water Database Table

```python
# Dependecies.
import pandas as pd
```

```python
# Get dataset from previous analysis.
cw_df = pd.read_csv('../resources/MAIN_country_happ_temp_water.csv', index_col=['Country', 'Year'])
cw_df.drop(columns=['Life Ladder', 'Temperature'], inplace=True)
cw_df.head()
```

```python
# Save table.
cw_df.dropna(inplace=True)
cw_df.to_csv('clean_water_table.csv')
```

```python
# Unstack also.
unstack_df = cw_df.unstack()
unstack_df = unstack_df.loc[:, 'Clean Water']
unstack_df.columns.name = None
unstack_df.to_csv('country_index/clean_water_table.csv')
unstack_df.head()
```
