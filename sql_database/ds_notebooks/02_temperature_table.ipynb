## Create the Temperature Database Table

```python
# Dependecies.
import pandas as pd
```

```python
# Import countries.
temp_df = pd.read_csv('../resources/M_countries_happiness_temps.csv', index_col='Country name')
temp_df.index.rename('Country', inplace=True)
temp_df = temp_df[['2010_TEMP', '2011_TEMP', '2012_TEMP', '2013_TEMP', '2014_TEMP', '2015_TEMP', '2016_TEMP', '2017_TEMP', '2018_TEMP', '2019_TEMP']]
temp_df.head()
```

```python
# Rename each column.
temp_df.columns = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
temp_df.head()
```

```python
# Save the database table.
temp_df.to_csv('country_index/temperature_table.csv')
temp_df.head()
```

```python
# Stack and save the database table.
stack_df = pd.DataFrame(temp_df.stack())
stack_df.index.set_names(['Country', 'Year'], inplace=True)
stack_df.rename(columns={0: 'Temperature'}, inplace=True)
stack_df.to_csv('temperature_table.csv')
stack_df.head()
```
