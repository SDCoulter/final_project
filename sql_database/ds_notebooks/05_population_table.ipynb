## Create the Population Database Table

```python
# Dependencies.
import pandas as pd
```

```python
# Greenhouse gas cleaned data.
pop_df = pd.read_csv('../resources/populations.csv')
pop_df.head()
```

```python
# Gather only needed columns.
pop_df = pop_df[['Country Name', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']]
pop_df.rename(columns={
    'Country Name': 'Country',
    '2010': 2010,
    '2011': 2011,
    '2012': 2012,
    '2013': 2013,
    '2014': 2014,
    '2015': 2015,
    '2016': 2016,
    '2017': 2017,
    '2018': 2018,
    '2019': 2019
}, inplace=True)
pop_df.head()
```

```python
# Fix country names.
pop_df.set_index('Country', inplace=True)
pop_df.rename(index={
    'Egypt, Arab Rep.': 'Egypt',
    'Kyrgyz Republic': 'Kyrgyzstan',
    'North Macedonia': 'Macedonia',
    'Russian Federation': 'Russia',
    'Slovak Republic': 'Slovakia',
    'Korea, Rep.': 'South Korea',
    'Venezuela, RB': 'Venezuela',
    'Yemen, Rep.': 'Yemen'
}, inplace=True)
pop_df.head()
```

```python
# Keep only those 100 countries we need.
countries = pd.read_csv('MAIN_table.csv', index_col='Country').index.unique().tolist()
pop_df_1 = pop_df.loc[pop_df.index.isin(countries)]
pop_df_1.head()
```

```python
# Sort index alphabetically.
pop_df_1 = pop_df_1.sort_index(axis=0)
pop_df_1.head()
```

```python
# Save DataFrame.
pop_df_1.to_csv('country_index/population_table.csv')
pop_df_1.head()
```

```python
# Stack and save the database table.
stack_df = pd.DataFrame(pop_df_1.stack())
stack_df.index.set_names(['Country', 'Year'], inplace=True)
stack_df.rename(columns={0: 'Population'}, inplace=True)
stack_df.to_csv('population_table.csv')
stack_df.head()
```
