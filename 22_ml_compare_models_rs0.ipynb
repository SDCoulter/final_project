Here we will read in the first two result DataFrames from notebooks 20 and 21, and compare how accurate they were.

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies.
import pandas as pd
```

```python
df_20 = pd.read_csv('ml_results/20_ml_results.csv')
df_20.head()
```

```python
df_21 = pd.read_csv('ml_results/21_ml_results.csv')
df_21.head()
```

```python
# Rename columns.
df_20.columns = ['20_Predicted', '20_Actual', '20_Difference']
df_21.columns = ['21_Predicted', '21_Actual', '21_Difference']
df_20.head()
```

```python
# Merge, and create difference DF.
merged_df = df_20.join(df_21)
diff_df = merged_df[['20_Difference', '21_Difference']]
diff_df.head()
```

```python
# Compare the differences.
diff_df.describe()
```

```python
# Style.
from matplotlib import style
style.use('fivethirtyeight')
```

```python
# Plot the differences.
bins = [0, 40, 80, 120, 160]
for i in bins:
    diff_df.iloc[i:i+40].plot(figsize=(15,3.5))
```

From the above graphs we are unable to make a clear distinction over which model is the better one, removing the country names and the years does not seem to affect the predicted life ladder scores.
