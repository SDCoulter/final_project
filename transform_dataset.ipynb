```python
# Dropping columns with null values
for column in air_df.columns:
    if air_df[column].isnull().sum() > 1:
        air_df.drop([column], axis=1, inplace=True)
```

```python
# Drop additional columns to make df cleaner
air_df.drop(['Indicator', 'IndicatorCode', 'Location type','ValueType', 'IndicatorCode','ParentLocationCode', 'ParentLocation', 'Period type', 'IsLatestYear', 'Dim1 type', 'Dim1ValueCode', 'DateModified', 'Language', 'FactValueNumericLow', 'FactValueNumericHigh', 'Value'], axis=1, inplace=True)
```

```python
# Rename columns to match another dataset
air_df = air_df.rename(columns={'SpatialDimValueCode': 'Code',
                        'Period': 'Year',
                        'Location': 'Country',
                        'Dim1': 'Location_type',
                        'FactValueNumeric': 'Air_quality'
                      })
```

```python
# Reorder columns for ease of use
air_df = air_df[['Country', 'Code', 'Year', 'Location_type', 'Air_quality']]
```

```python
# Sort columns by countries and ascending
air_df.sort_values(by=['Country', 'Year'], inplace=True)
```

```python
# Reset the index of df
air_df = air_df.reset_index(drop=True)
air_df.head()
```

```python
for column in water_df.columns:
    if water_df[column].isnull().sum() > 4000:
        water_df.drop([column], axis=1, inplace=True)
```

```python
# Drop additional columns to make df cleaner
water_df.drop(['Indicator', 'IndicatorCode', 'Location type','ValueType', 'IndicatorCode','ParentLocationCode', 'ParentLocation', 'Period type', 'IsLatestYear', 'Dim1 type', 'Dim1ValueCode', 'DateModified', 'Language', 'Value'], axis=1, inplace=True)
```

```python
# Fill NA for ease of use
water_df.fillna(0, inplace=True)

# Rename columns to match another dataset
water_df.rename(columns={'SpatialDimValueCode': 'Code',
                    'Period': 'Year',
                    'Location': 'Country',
                    'Dim1': 'Location_type',
                    'FactValueNumeric': 'Water_quality'
                  }, inplace=True)
```

```python
# Reorder columns for ease of use
water_df = water_df[['Country', 'Code', 'Year', 'Location_type', 'Water_quality']]
```

```python
water_df = water_df[water_df['Year'] >= 2008]
```

```python
# Sort columns by countries and ascending
water_df.sort_values(by=['Country', 'Year'], inplace=True)

# Reset the index of df
water_df = water_df.reset_index(drop=True)
```

```python
air_water_df = water_df.merge(air_df, how='left')
```

```python
air_water_df.head()
```

```python
air_water_df.head()
```

```python
# Show only 'Total' data without 'Rural' and 'Urban'
air_water_total_df = air_water_df[air_water_df['Location_type'] == 'Total']
air_water_total_df.drop(columns='Location_type', axis=1, inplace=True)

# Reset the index of df
air_water_total_df = air_water_total_df.reset_index(drop=True)
```

```python
# Check merged data
air_water_total_df.head(25)
```

```python
# export merged data
air_water_total_df.to_csv('resources/air_water.csv')
```
