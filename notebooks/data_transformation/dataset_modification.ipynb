```python
import pandas as pd
```

```python
co2_df = pd.read_csv('resources/co2_emission.csv')
co2_df.head()
```

```python
# only show data from 2008 and onwards
co2_df = co2_df[co2_df['Year'] >= 2008]
co2_df.head(20)
```

```python
happiness_df = pd.read_csv('resources/world-happiness-report.csv')
happiness_df.head(10)
```

```python
# Matching both dataframes' column names
happiness_df = happiness_df.rename(columns={"year": "Year", "Country name": "Country"})
co2_df = co2_df.rename(columns={"Entity": "Country"})
```

```python
happiness_df.head(20)
```

```python
# merge two df
df = pd.merge(co2_df, happiness_df,  how='left', left_on=['Country','Year'], right_on = ['Country','Year'])
```

```python
df.head(20)
```

```python
# check number of unique countries
len(co2_df['Country'].unique())
```

```python
# check number of unique countries
len(happiness_df['Country'].unique())
```

```python
df.isnull().sum()
```

```python
# since a lot of country data are missing, countries that do not have happiness data are dropped
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# renaming columns for ease of use
df = df.rename(columns={"Annual CO₂ emissions (tonnes )": "Annual_CO2_emission_tonnes",
                        "Life Ladder": "Life_ladder",
                        "Log GDP per capita": "Log_GDP_per_capita",
                        "Social support": "Social_support",
                        "Healthy life expectancy at birth": "Life_expectancy_at_birth",
                        "Freedom to make life choices": "Freedom_to_make_life_choices",
                        "Perceptions of corruption": "Perceptions_of_corruption",
                        "Positive affect": "Positive_affect",
                        "Negative affect": "Negative_affect"
                       })
```

```python
# check merged and modified df
df.head(10)
```

```python
# # export df to csv file
df.to_csv('resources/co2_happiness_modified.csv')
```
