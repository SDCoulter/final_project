```python
import pandas as pd
```

```python
# read main datset
df = pd.read_csv('../../dataset/MAIN_happ_temp_cw_gg_pop_pm.csv')
```

```python
df.head(10)
```

```python
# drop 2021 data
df = df[df['Year'] != 2021]
```

```python
# groupby year and country for potential use
g_df = df.groupby(by=['Year', 'Country']).sum()
g_df
```

```python
# with null values in 2019 on greenhosue gas emissions, 2018 is used for calculations
change_df = df[(df['Year'] == 2010) | (df['Year'] == 2018)]
change_df.reset_index(drop=True, inplace=True)
change_df
```

```python
# country and year columns are not needed and therefore removed
# then use diff() for every row to get the change between each row
change_df = change_df[['Life Ladder','Temperature','Clean Water','Greenhouse Gas Emissions','Population','PM2.5']].diff(periods=1)
```

```python
# first row is null value since it has nothing to subtract, dropped and reset index here
change_df = change_df[1:]
change_df.reset_index(drop=True)
```

```python
# slice df every 2 rows since calculation was 2018-2010, therefore even rows aren't needed here
change_df = change_df.iloc[::2]
change_df.head()
```

```python
# reset index again
change_df.reset_index(drop=True, inplace=True)
```

```python
# get the country column from original df and add to the df with calculation
change_df['Country'] = df['Country'].unique()
```

```python
# rearrange columns
change_df = change_df[['Country', 'Life Ladder', 'Temperature', 'Clean Water', 'Greenhouse Gas Emissions', 'Population', 'PM2.5']]
```

```python
# check df
change_df.head()
```

```python
# can be exported and used for visualization
change_df.to_csv('../../dataset/change_2010-2018.csv', index=False)
```
