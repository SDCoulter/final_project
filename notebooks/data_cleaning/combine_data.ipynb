```python
import pandas as pd
```

```python
df = pd.read_csv('Resources/MAIN_country_happ_temp_water.csv')
df
```

```python
df2 = pd.read_csv('Resources/greenhouse_gasses_clean.csv')
df2
```

```python
merged_df = pd.merge(df, df2,  how='left', left_on=['Country','Year'], right_on = ['Country Name','Date'])
clean_df = merged_df.drop(columns=['Country Name', 'Date'])
clean_df
```

```python
final_df = clean_df.rename(columns={"Value": "Greenhouse Gas Emissions"})
final_df
```

```python
final_df.to_csv('resources/MAIN_country_happ_temp_water_gg.csv', index=False)
```
